#### Problema 3:
Un número al que se le resta 7 da 18Paso 1: Análisis y ResoluciónPlanteamiento: Sea $x$ el número desconocido.Ecuación: "Un número ($x$) menos 7 ($- 7$) es igual a 18".$$x - 7 = 18$$Despeje: El 7 está restando, por lo que pasa al otro lado sumando.$$x = 18 + 7$$Solución:$$x = 25$$

```python
# Problema 3: Despeje con resta
resultado = 18
resta = 7

# x = resultado + resta
x = resultado + resta
print(f"El número es: {x}")
```

#### Problema 4:
El triple de un número más 4 es igual a 31Paso 1: Análisis y ResoluciónPlanteamiento: "El triple de un número" se representa como $3x$.Ecuación:$$3x + 4 = 31$$Despeje (paso a paso):Primero quitamos lo que suma: El 4 pasa restando.$$3x = 31 - 4$$$$3x = 27$$Luego quitamos lo que multiplica: El 3 pasa dividiendo.$$x = \frac{27}{3}$$Solución:$$x = 9$$

```python
# Problema 4: Ecuación de dos pasos
total = 31
suma = 4
multiplicador = 3

# Primero restamos, luego dividimos
x = (total - suma) / multiplicador
print(f"El número es: {int(x)}")
```

 #### Problema 5:
 Terreno donde el largo mide 5m más que el anchoNota importante: Para resolver este problema necesitamos un dato adicional (como el perímetro o el área). Asumiendo que te refieres al planteamiento base de las dimensiones:

 Paso 1: Análisis y ResoluciónPlanteamiento:
  * Ancho: $a$Largo:
  $a + 5$Ecuación:
  
   Si buscamos representar las dimensiones, la relación es:$$Largo = a + 5$$Análisis: Si el problema tuviera un perímetro de 50 metros (por ejemplo), la ecuación sería:

 $2(a) + 2(a + 5) = 50$.$2a + 2a + 10 = 50$$4a = 40$$a = 10$Solución (con perímetro 50m): El ancho es 10m y el largo es 15m.

```python
# Problema 5: Dimensiones del terreno (ejemplo con perímetro = 50)
perimetro = 50

# Ecuación: 4a + 10 = perimetro -> a = (perimetro - 10) / 4
ancho = (perimetro - 10) / 4
largo = ancho + 5

print(f"Ancho: {ancho}m, Largo: {largo}m")
```
