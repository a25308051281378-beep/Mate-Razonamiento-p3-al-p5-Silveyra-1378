#### Problema 3:
Un número al que se le resta 7 da 18Paso 1: Análisis y ResoluciónPlanteamiento: Sea $x$ el número desconocido.Ecuación: "Un número ($x$) menos 7 ($- 7$) es igual a 18".$$x - 7 = 18$$Despeje: El 7 está restando, por lo que pasa al otro lado sumando.$$x = 18 + 7$$Solución:$$x = 25$$

```python
# Problema 3: Despeje con resta
resultado = 18
resta = 7

# x = resultado + resta
x = resultado + resta
print(f"El número es: {x}")
```

#### Problema 4:
El triple de un número más 4 es igual a 31Paso 1: Análisis y ResoluciónPlanteamiento: "El triple de un número" se representa como $3x$.Ecuación:$$3x + 4 = 31$$Despeje (paso a paso):Primero quitamos lo que suma: El 4 pasa restando.$$3x = 31 - 4$$$$3x = 27$$Luego quitamos lo que multiplica: El 3 pasa dividiendo.$$x = \frac{27}{3}$$Solución:$$x = 9$$

```python
# Problema 4: Ecuación de dos pasos
total = 31
suma = 4
multiplicador = 3

# Primero restamos, luego dividimos
x = (total - suma) / multiplicador
print(f"El número es: {int(x)}")
```

#### problema 5: Geometría y Perímetro
Enunciado: Un terreno tiene un largo que mide 5 metros más que el ancho. Si el perímetro total es de 50 metros, ¿cuánto mide cada lado?Paso 1: Análisis y PlanteamientoDefinir variables:Ancho = $w$Largo = $w + 5$ (porque mide 5 metros más que el ancho).Fórmula del perímetro: El perímetro es la suma de todos los lados de un rectángulo (2 anchos y 2 largos):$$Perímetro = 2(ancho) + 2(largo)$$Sustituir con nuestros datos:$$50 = 2(w) + 2(w + 5)$$Paso 2: Resolución (Despeje)Eliminar paréntesis: Multiplicamos el 2 por lo que está dentro:$$50 = 2w + 2w + 10$$Agrupar términos semejantes: Sumamos las $w$:$$50 = 4w + 10$$Despejar $w$:Pasamos el 10 restando: $50 - 10 = 4w$ $\Rightarrow$ $40 = 4w$.Pasamos el 4 dividiendo: $w = \frac{40}{4}$ $\Rightarrow$ $w = 10$.Encontrar dimensiones:Ancho: 10 metros.Largo: $10 + 5 = 15$ metros

```python
# Problema 5: 2(w + 5) + 2w = 50 -> 4w + 10 = 50
# perimetro = 50
# diferencia_largo = 5
perimetro=int(input("ingrese el perimetro: "))
diferencia_largo=int(input("ingrese la diferencia de largo: "))
# Despeje: w = (50 - (2 * 5)) / 4
ancho = (perimetro - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"Ancho del terreno: {ancho} m")
print(f"Largo del terreno: {largo} m")
```
